## 최적의 유성우 관측일

```python
cities
```

```python
def predict_best_meteor_shower_viewing(city):
    latitude = cities.loc[cities['city'] == city, 'latitude'].iloc[0]
    constellations_list = constellations.loc[(latitude<=constellations['latitudestart']) & (latitude>=constellations['latitudeend']), 'constellation'].tolist()
    if constellations_list:
        meteor_shower_string = city + '에서는 별자리 관측이 가능합니다'
    else:
        meteor_shower_string = city + '에 관한 데이터가 없습니다'
    
    return latitude, constellations_list, meteor_shower_string
```

```python
# 도시명을 입력하면 관측가능한 유성우에 대한 정보를 알려주는 함수 작성하기

def predict_best_meteor_shower_viewing(city):
    # 안내 메시지 초기화
    meteor_shower_string = ''

    # 입력한 도시 정보가 존재하지 않으면 오류 메시지
    if(city not in cities['city'].values):
        return (f'도시 정보를 찾을 수 없습니다.\n\t도시: {city}')
    
    # 도시의 위도 정보를 불러오기
    latitude = cities.loc[cities['city'] == city, 'latitude'].iloc[0]

    # 해당 도시의 위도에서 관측 가능한 별자리 조회
    constellations_list = constellations.loc[(latitude <= constellations['latitudestart']) & (latitude >= constellations['latitudeend']), 'constellation'].tolist()
    
    # 별자리 관측 가능 안내 메시지
    if constellations_list:
        meteor_shower_string = city + '에서는 별자리 관측이 가능합니다'
    # 볼 수 있는 별자리가 존재하지 않으면 오류 메시지
    else:
        return (f'관측 가능한 별자리가 없습니다.\n\t도시: {city}')

    # 도시별 관찰 가능한 별자리 iterate
    for constellation in constellations_list:
        # 별자리와 가장 가까운 유성우 조회
        meteor = meteor_showers[meteor_showers['radiant'] == constellation].iloc[0]
        print(meteor)

        # 유성우 관측이 가능한 시작일과 종료일
        startdate = meteor['startdate']
        enddate = meteor['enddate']

        # 유성우가 보일 때의 달의 위상 조회
        moon_phase_list = moon_phases[(moon_phases['date'] >= startdate) & (moon_phases['date'] <= enddate)]

        # 가장 어두운 날짜를 idxmin()으로 찾기
        best_moon_date = moon_phase_list.loc[moon_phase_list['percentage'].idxmin()]['date']
        darkest_date_first = best_moon_date.date()  # 'date()'로 Timestamp에서 날짜만 추출

        # 사용자에게 정보전달
        meteor_shower_string += f"{meteor['name']} 자리를 잘 보려면 {meteor['radiant']} 자리를 잘 보아야 합니다.\n{str(darkest_date_first)}부터 잘 관측 가능합니다.\n"

    return meteor_shower_string
```
