# 데이터 불러오기
* **`pd.read_csv(csv파일 경로, parse_dates=[datetime형식으로 불러올 컬럼])`**

* 속성 정보
    * `datetime` - 시간대별 데이터 제공. timestamp(날짜와 시간 결합) 형태
    * `season`  - 1 = spring, 2 = summer, 3 = fall, 4 = winter
    * `holiday` - 공휴일(주말이 아닌 빨간날). 1-yes, 2-no
    * `workingday` - 주말, 공휴일을 제외한 날 1-yes, 2-no
    * `weather`
        * 1: Clear, Few clouds, Partly cloudy, Partly cloudy
        * 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
        * 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
        * 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
    * `temp` - 섭씨온도
    * `atemp` - 섭씨체감온도
    * `humidity` - 상대습도
    * `windspeed` - 풍속
    * `casual` - 비회원 대여량
    * `registered` - 회원 대여량
    * `count` - 총 대여량


```python
# datetime 컬럼은 datetime 형식으로 읽어오도록 옵션 설정
df = pd.read_csv('https://raw.githubusercontent.com/JayoungKim-ai/ML_dataset/main/BikeSharingDemand.csv', parse_dates=['datetime'])
# df = pd.read_csv('data/bike_sharing_demand.csv', parse_dates=['datetime'])
display(df.head(3))
display(df.tail(3))
```

```python
# 데이터 정보
df.info()
```

# 파생컬럼 추가

```python
# 연, 월, 일, 시, 요일
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['dayofweek'] = df['datetime'].dt.dayofweek # 월요일:0, 일요일:6
df.head()
```

# LinearRegression

```python
# 변수선택
x = df[['temp', 'humidity', 'season', 'holiday', 'workingday', 'weather', 'windspeed', 'year', 'month', 'day', 'hour', 'dayofweek']]
y = df['count']
```

```python
# 훈련세트/테스트세트 분할
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
# 모델 생성
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# 모델 훈련
model.fit(x_train, y_train)
```

```python
# 테스트데이터로 예측
# 테스트데이터 예측값
pred = model.predict(x_test)

# 모델 성능 평가
# 정답과 예측값값
# 모델 평가를 위한 함수
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, pred)
rmse = root_mean_squared_error(y_test, pred)
mae = mean_absolute_error(y_test, pred)
r2 = r2_score(y_test, pred)
print(f'MSE : {mse:.2f}')
print(f'RMSE: {rmse:.2f}')
print(f'MAE : {mae:.2f}')
print(f'R2 : {r2:.2f}')
```

## 원핫인코딩

```python
# 범주형변수 원핫인코딩
x = df[['temp', 'holiday', 'workingday', 'windspeed', 'season', 'weather', 'windspeed', 'year', 'month', 'day', 'hour', 'dayofweek']]
# y = df['count']

x_dummi = pd.get_dummies(x, columns=['year', 'month', 'day', 'hour', 'dayofweek', 'season', 'weather'])

x_dummi.head()
```

```python
# 훈련세트/테스트세트 분할
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_dummi, y, test_size= 0.2, random_state=1)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

# 모델 생성
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# 모델 훈련
model.fit(x_train, y_train)

# 모델 성능 평가
# 예측값값
pred = model.predict(x_test)

# 모델 성능 평가
# 정답과 예측값값
# 모델 평가를 위한 함수
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, pred)
rmse = root_mean_squared_error(y_test, pred)
mae = mean_absolute_error(y_test, pred)
r2 = r2_score(y_test, pred)
print(f'MSE : {mse:.2f}')
print(f'RMSE: {rmse:.2f}')
print(f'MAE : {mae:.2f}')
print(f'R2 : {r2:.4f}')
```

## 종속변수 로그변환

```python
y.hist()
```

```python
# x 전처리
# # 범주형변수 원핫인코딩
x = df[['temp', 'holiday', 'workingday', 'windspeed', 'season', 'weather', 'windspeed', 'year', 'month', 'day', 'hour', 'dayofweek']]
# y = df['count']
x_dummi = pd.get_dummies(x, columns=['year', 'month', 'day', 'hour', 'dayofweek', 'season', 'weather'])

#y 전처리
# 종속 변수 로그 변환환
y_log = np.log1p(y)
y_log.hist()
```

로그 역변환X

```python
# 훈련세트/테스트세트 분할
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_dummi, y_log, test_size= 0.2, random_state=1)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

# 모델 생성
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# 모델 훈련
model.fit(x_train, y_train)

# 모델 성능 평가
# 예측값 - 로그 변환이 된 값으로 했을 것임 -> 로그 역변환을 해주어야 함
pred = model.predict(x_test)


# 모델 성능 평가
# 정답과 예측값
# 모델 평가를 위한 함수
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, pred)
rmse = root_mean_squared_error(y_test, pred)
mae = mean_absolute_error(y_test, pred)
r2 = r2_score(y_test, pred)
print(f'MSE : {mse:.2f}')
print(f'RMSE: {rmse:.2f}')
print(f'MAE : {mae:.2f}')
print(f'R2 : {r2:.4f}')
```

로그 역변환

```python
# 훈련세트/테스트세트 분할
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_dummi, y_log, test_size= 0.2, random_state=1)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

# 모델 생성
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# 모델 훈련
model.fit(x_train, y_train)

# 모델 성능 평가
# 예측값 - 로그 변환이 된 값으로 했을 것임 -> 로그 역변환을 해주어야 함함
pred = model.predict(x_test)

# 로그 역변환
pred_relog = np.expm1(pred)
y_test_relog = np.expm1(y_test)

# 모델 성능 평가
# 정답과 예측값
# 모델 평가를 위한 함수
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test_relog, pred_relog)
rmse = root_mean_squared_error(y_test_relog, pred_relog)
mae = mean_absolute_error(y_test_relog, pred_relog)
r2 = r2_score(y_test_relog, pred_relog)
print(f'MSE : {mse:.2f}')
print(f'RMSE: {rmse:.2f}')
print(f'MAE : {mae:.2f}')
print(f'R2 : {r2:.4f}')
```

## cross validation

```python
# model 새로 생성
model = LinearRegression()

from sklearn.model_selection import KFold, cross_val_score

# KFold 객처 생성
kf = KFold(n_splits=5, shuffle=True, random_state=1)
print(cross_val_score(model, x, y, scoring='r2', cv=kf).mean())
# 원핫 인코딩 한거
print(cross_val_score(model, x_dummi, y, scoring='r2', cv=kf).mean())
# 원핫 인코딩, y 로그 변환
print(cross_val_score(model, x_dummi, y_log, scoring='r2', cv=kf).mean())
```
